## El detector

El detector tiene una apertura vertical (rendija) que, cuando está centrada ($A_c=0$), subtiende un semi-ángulo $\alpha_0$ desde el origen.

La semi-altura física de la rendija es $h = D \tan(\alpha_0)$.

Cuando el detector se desplaza verticalmente una distancia $A_c$, la rendija cubre posiciones verticales desde $A_c - h$ hasta $A_c + h$. Un electrón con ángulo $\theta$ llega a la posición vertical $y = D\tan(\theta)$, y pasa el filtro si $A_c - h \leq y \leq A_c + h$.

```python
# Descomentar cuando se ejecuta desde la página web
import piplite
await piplite.install(['ipywidgets'])

import numpy as np
import matplotlib.pyplot as plt

D_DETECTOR = 5.0  # distancia horizontal del detector al origen

def detector(momentos, alpha_0, Ac=0.0):
    """
    Filtra los momentos que pasan por la rendija del detector.

    El detector está a distancia horizontal D=5 del origen.
    La rendija tiene semi-altura h = D*tan(alpha_0) y está
    centrada en la posición vertical Ac.

    Parámetros
    ----------
    momentos : ndarray de forma (N, 3)
        Cada fila contiene [|p|, theta, E_cin], donde:
        - |p|    : módulo del momento
        - theta  : ángulo respecto al eje horizontal (rad)
        - E_cin  : energía cinética (|p|^2 / 2, con m=1)
    alpha_0 : float
        Semi-apertura angular del detector (rad).
    Ac : float
        Desplazamiento vertical del centro de la rendija.

    Retorna
    -------
    ndarray de forma (M, 3)
        Subconjunto de filas de `momentos` que pasan el filtro.
    """
    # h = D_DETECTOR * np.tan(alpha_0)
    # y = D_DETECTOR * np.tan(momentos[:, 1])
    # mascara = (y >= Ac - h) & (y <= Ac + h) & (np.abs(momentos[:,1])<np.pi/2)
    px=momentos[:,0]*np.cos(momentos[:,1])
    py=momentos[:,0]*np.sin(momentos[:,1])
    nuevo_py=py+Ac
    nuevo_theta=np.atan2(nuevo_py,px)
    nuevo_p=np.sqrt(px**2+nuevo_py**2)
    nueva_Ec=nuevo_p**2/2
    nuevos_momentos=np.vstack((nuevo_p,nuevo_theta,nueva_Ec)).T    
    mascara = np.abs(nuevos_momentos[:,1])<alpha_0

    return nuevos_momentos[mascara]
```
